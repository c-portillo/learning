## Gradient Descent

```python
# gradient descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
gradient = 0

x_list = []

for i in range(500):
    previous_x = new_x
    gradient = df(new_x)
    new_x = previous_x - (gradient * step_multiplier)
    x_list.append(new_x)
    
print('Local Minima: ', new_x)
print('Gradient: ', df(new_x))
```

# Example 2: Multiple Minima vs Initial Guess and Advanced Functions

## $$g(x) = x^4 - 4x^2 + 5$$

```python
# generate data
x_2 = np.linspace(-2, 2, 1000) # generate 1000 datapoints between -2 and 2

# g(x) function 
def g(x):
    return x**4 - 4*x**2 + 5

# dg(x) function 
def dg(x):
    return 4*x**3 - 8*x

```

# Gradient Descent as a Python Function

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.00001, max_iter=500): # returns: new_x, x_list, slope_list

    # gradient descent
    new_x = initial_guess

    x_list = [new_x]

    for i in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - (gradient * multiplier)
        
        x_list.append(new_x)

        step_size = abs(new_x - previous_x)
        if step_size < precision: 
            break
            
    return new_x, x_list    # we are returning a tuple
```

```python
local_min, list_x = gradient_descent(derivative_func=dg, initial_guess=0)

# plot function and derivative side by side
plt.figure(figsize=(15, 5))

# 1 Chart: Cost Function
plt.subplot(1, 2, 1)    # subplot(row, column, index)
plt.title('Cost Function', fontsize=20)
plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3)
plt.plot(local_min, g(local_min), 'ro', markersize=10) 

# 2 Chart: Derivative
plt.subplot(1, 2, 2)    # subplot(row, column, index)
plt.title('Slope of Cost Function', fontsize=20)
plt.xlim(-2, 2)
plt.ylim(-6, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5)
plt.plot(local_min, dg(local_min), 'ro', markersize=10)
plt.grid()    # add grid


plt.show()
```

# How do we choose a proper learning rate?

```python
n = 100  # max iterations 

# store in tuple
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.0005, max_iter=n)

# plot reduction in cost for each iteration 
plt.figure(figsize=(20, 10))

plt.ylim(0, 50)
plt.xlim(0, n)

plt.title('Effect of Learning Rate', fontsize=20)
plt.ylabel('Cost', fontsize=15)
plt.xlabel('Number Iterations', fontsize=15)

# values for chart

# y-axis data
low_values = np.array(low_gamma[1])

# x-axis data: a list from 0 to n+1 (the number of iterations plus initial guess)
iteration_list = list(range(n+1)) # create list of values from 0 to n+1


# Plotting low learnring rate (Green)
plt.plot(iteration_list, g(low_values), color='lightgreen', linewidth=5)
plt.scatter(iteration_list, g(low_values), color='lightgreen', s=80)


mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.001, max_iter=n)
mid_values = np.array(mid_gamma[1])

# Plotting medium learning rate (Yellow)
plt.plot(iteration_list, g(mid_values), color='yellow', linewidth=5)
plt.scatter(iteration_list, g(mid_values), color='yellow', s=80)

high_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.002, max_iter=n)
high_values = np.array(high_gamma[1])

# Plotting high learning rate (Orange)
plt.plot(iteration_list, g(high_values), color='orange', linewidth=5)
plt.scatter(iteration_list, g(high_values), color='orange', s=80)

insane_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9, multiplier=0.25, max_iter=n)
insane_values = np.array(insane_gamma[1])

# Plotting insane learning rate (Red)
plt.plot(iteration_list, g(insane_values), color='red', linewidth=5)
plt.scatter(iteration_list, g(insane_values), color='red', s=80)
```

### As we can see above, there can be advantages to having higher learning rates. The graph above shows the effects of learning rates from smallest largest (green, yellow, orange, red, in that order). However, the learning rate can be too large cause the algorithm to end up in an infinite loop, just bouncing around all over the place (as shown by the red line)

### Another possible approach to learning rate is implementing the gradient descent algorithm with a learning that updates by decreasing after each iteration. The idea behind this is that the further you are from the minima, the larger the step you have to take. As you move closer, the step should be smaller (because we don't want to overshoot) and thus, we decrease the learning rate. 

### Yet another possible approach to learning rate is one called the "Bold Driver." It works as follows: If error rate decreased from previous iteration (moved closer to minima), then increase learning rate by 5%, else if error rate increased (overshot the minima), then we reset to the previous iteration and decrease the learning rate by 50%.

### There are many approaches to setting the learning rate!

## Partial Derivatives and Symbolic Computation

## Batch Gradient Descent with SymPy

```python
# setup
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0]) # initial guess

for i in range(max_iter):
    gradient_x = diff(f(a, b), a).evalf(subs={a:params[0], b:params[1]})
    gradient_y = diff(f(a, b), b).evalf(subs={a:params[0], b:params[1]})
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier*gradients
    
print('Values in gradient array: ', gradients)
print('Minimum occurs at x value of: ', params[0])
print('Minimum occurs at y value of: ', params[1])
print('The cost is: ', f(params[0], params[1]))
```

### Partial derivatives as Python functions

```python
# partial derivative with respect to x
def fpx(x, y):
    r = 3**(-x**2 -y**2)
    return 2 * r * x * math.log(3) / (r + 1)**2

# partial derivative with respect to y
def fpy(x, y):
    r = 3**(-x**2 -y**2)
    return 2 * r * y * math.log(3) / (r + 1)**2
```

## Batch Gradient Descent using our Python Functions (same results)

```python
# setup
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0]) # initial guess

for i in range(max_iter):
    gradient_x = fpx(params[0], params[1])
    gradient_y = fpy(params[0], params[1])
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier*gradients
    
print('Values in gradient array: ', gradients)
print('Minimum occurs at x value of: ', params[0])
print('Minimum occurs at y value of: ', params[1])
print('The cost is: ', f(params[0], params[1]))
```

### Note! Gradient Descent with python functions we wrote runs faster than the previous one using SymPy.  SymPy might be simpler for finding derivatives, but it can also be more computationally intensive. 

# Graphing 3D Gradient Descent

```python
# setup
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0]) # initial guess
values_array = params.reshape(1, 2) # 1 row, 2 columns

for i in range(max_iter):
    gradient_x = fpx(params[0], params[1])
    gradient_y = fpy(params[0], params[1])
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier*gradients
    #values_array = np.append(arr=values_array, values=params.reshape(1, 2), axis=0)
    values_array = np.concatenate((values_array, params.reshape(1,2)), axis=0) # a different way to do the line above
    

# print(values_array.shape)
# print(values_array)
print('Values in gradient array: ', gradients)
print('Minimum occurs at x value of: ', params[0])
print('Minimum occurs at y value of: ', params[1])
print('The cost is: ', f(params[0], params[1]))
```

# Example 5: Working with Data & a Real Cost Function

## Mean Squared Error: A cost function for regression problems

### $$ RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta(x^{(i)}) \big)^2$$
### $$ MSE = \frac{1}{n}\sum_{i=1}^{n} \big( y^{(i)} - h_\theta(x^{(i)}) \big)^2$$
### $$ MSE = \frac{1}{n}\sum_{i=1}^{n} \big( y^{(i)} - \hat{y} \big)^2$$

#### MSE is more scalable than RSS, is it prevents the number from getting too large which can prevent overflow errors

```python
# make sample data
x_5 = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose() # add extra brackets, and transpose 
y_5 = np.array([1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]).reshape(7, 1) # reshape(rows, cols)
# note 2 different ways to reshape np array above

# for linear regression to work, x and y must be vectors
print(x_5.shape)
print(y_5.shape)
```

```python
# Quick Linear Regression
regr = LinearRegression();
regr.fit(x_5, y_5)

print('theta_0 (intercept): ', regr.intercept_[0])
print('theta_1 (slope): ', regr.coef_[0][0])
```

```python
plt.scatter(x_5, y_5, s=50)
plt.plot(x_5, regr.predict(x_5), color='r')
```

### Let's find the Mean Squared Error (MSE)

### Recall: $$ MSE = \frac{1}{n}\sum_{i=1}^{n} \big( y^{(i)} - \hat{y} \big)^2$$

```python
# y_hat = theta_0 + theta_1 * x_5
y_hat = 0.8475351486029536 + 1.2227264637835915 * x_5
print('Estimated y_hat values: \n', y_hat)
print('Actual y values: \n', y_5)
```

```python
# MSE
def MSE(y_vals, y_hat):
    mse_calc = (np.sum((y_vals - y_hat)**2))/len(y_vals)
#     mse_calc = (sum((y_vals - y_hat)**2))/len(y_hat)
#     mse_calc = np.average((y_vals - y_hat)**2, axis=0) # axis=0 sums up rows, axis=1 sums up cols
    
    return mse_calc
```

```python
print('our MSE function:     ', MSE(y_5, y_hat))
print('sklearn MSE function: ', mean_squared_error(y_5, y_hat))
```
